```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("tab.csv",skiprows=[0,2],usecols=range(2,31))
ar=[]
names=data.columns
for name in data.columns:
    g=data[name]
    g=pd.to_numeric(g, errors='coerce')
    g = g[np.isnan(g)]
    ar.append(g.size)
data=np.array(ar);
plt.subplots(figsize=(100,15)) 
plt.bar(names, data, align='center')
plt.title('Missing Values')
plt.savefig("miss_val.jpg")
```
