```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("corrMatrix.csv",header=None);
names = pd.read_csv("tab.csv",skiprows=1,usecols=range(2,31));
data=pd.DataFrame(data.values,columns=names.columns,index=names.columns)
fig, ax = plt.subplots(figsize=(25,25)) 
ax = sns.heatmap(data, annot=True, linewidths=.5, ax=ax);
fig.savefig("heatmap.jpg")
```
